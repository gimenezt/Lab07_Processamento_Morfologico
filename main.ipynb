```python
#3.b
def gradiente_morf(img, ksize):
    k = cv2.getStructuringElement(cv2.MORPH_RECT, (ksize, ksize))
    return cv2.absdiff(cv2.dilate(img, k), cv2.erode(img, k))

def webcam_morfologico():
    cap = cv2.VideoCapture(0)
    if not cap.isOpened():
        print("Erro: webcam nao encontrada")
        return

    ksize = 3
    gravando = False
    out = None

    print("GRADIENTE MORFOLOGICO")
    print("1-9: muda kernel (1=3x3 ate 9=19x19)")
    print("r: gravar video | s: screenshot | q: sair")

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        _, binaria = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
        grad = gradiente_morf(binaria, ksize)
        grad_bgr = cv2.cvtColor(grad, cv2.COLOR_GRAY2BGR)

        info = f"Kernel: {ksize}x{ksize}"
        cv2.putText(frame, info, (10,30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,255,0), 2)
        cv2.putText(grad_bgr, info, (10,30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255,255,255), 2)

        if gravando:
            cv2.putText(frame, "REC", (10,60), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,0,255), 2)

        resultado = cv2.hconcat([frame, grad_bgr])
        cv2.imshow('Gradiente Morfologico - Original vs Extraido', resultado)

        if gravando and out:
            out.write(resultado)

        key = cv2.waitKey(1) & 0xFF
        if key == ord('q'):
            break
        elif key == ord('s'):
            cv2.imwrite(f'screenshot_k{ksize}.png', resultado)
            print(f"Screenshot salvo")
        elif key == ord('r'):
            gravando = not gravando
            if gravando:
                h,w = resultado.shape[:2]
                out = cv2.VideoWriter('webcam_morfologico.avi', cv2.VideoWriter_fourcc(*'XVID'), 20, (w,h))
                print("Gravando...")
            else:
                if out:
                    out.release()
                    out = None
                print("Gravacao parada")
        elif ord('1') <= key <= ord('9'):
            ksize = (key - ord('0')) * 2 + 1
            print(f"Kernel: {ksize}x{ksize}")

    if out:
        out.release()
    cap.release()
    cv2.destroyAllWindows()
```

```python

#4.a

def processar_para_ocr(img_bin):
    # Remove ruido pequeno (abertura)
    k_peq = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
    sem_ruido = cv2.morphologyEx(img_bin, cv2.MORPH_OPEN, k_peq)
    # Conecta caracteres (fechamento)
    k_med = cv2.getStructuringElement(cv2.MORPH_RECT, (5,5))
    conectado = cv2.morphologyEx(sem_ruido, cv2.MORPH_CLOSE, k_med)
    return sem_ruido, conectado

print("\nProcessando imagem do grupo:")
s_ruido, conect = processar_para_ocr(grupo_bin)

fig, axes = plt.subplots(1, 4, figsize=(16, 4))
axes[0].imshow(grupo_bin, cmap='gray')
axes[0].set_title('Original binarizada')
axes[0].axis('off')
axes[1].imshow(s_ruido, cmap='gray')
axes[1].set_title('Abertura (remove ruido)')
axes[1].axis('off')
axes[2].imshow(conect, cmap='gray')
axes[2].set_title('Fechamento (conecta)')
axes[2].axis('off')
axes[3].imshow(cv2.absdiff(grupo_bin, s_ruido), cmap='gray')
axes[3].set_title('Ruido removido')
axes[3].axis('off')
plt.tight_layout()
plt.show()
print(f"Pixels de ruido removidos: {np.sum(cv2.absdiff(grupo_bin, s_ruido) > 0)}")
```

Analise final

BINARIZACAO:
- Fundamental para morfologia. Metodo de Otsu foi mais eficaz.
- Binarizacao manual (threshold fixo) perde detalhes.

TAMANHO DO KERNEL:
- 3x3: preserva detalhes, remove ruido pequeno
- 5x5-7x7: equilibrio entre filtragem e preservacao
- 11x11+: altera significativamente a estrutura, remove detalhes importantes

FORMATO DO KERNEL:
- Retangular: preserva cantos e bordas retas
- Eliptico: bordas mais suaves, arredondadas
- Cruz: preserva apenas linhas horizontais/verticais

APLICACOES PRATICAS:
- OCR: abertura remove poeira, fechamento conecta letras
- Segmentacao: gradiente morfologico detecta bordas
- Remocao de ruido: erosao+dilatacao (abertura)
- Preenchimento: dilatacao+erosao (fechamento)

ARTEFATOS OBSERVADOS:
- Erosao excessiva apaga objetos pequenos
- Dilatacao excessiva funde objetos vizinhos
- Kernel muito grande causa perda de informacao
